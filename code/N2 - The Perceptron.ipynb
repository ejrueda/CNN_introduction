### Perceptron

- La unidad básica, es la representación de una neurona biológica

```python
Image.open("../images/perceptron.png")
```

- donde $x_1,\dots,x_n$ representa el vector de caracteristicas de entrada y el <b>bias</b> representa una constante de entrada.
- $w_0,\dots,w_n$ representan los pesos del perceptron, los cuales mediante un algoritmo de aprendizaje serán ajustados para obtener el menor error posible.
- $h$ es la función de salida antes de pasar por la función de activación. Esta viene dada por:
$$ h =  w_1x_1 + w_2x_2 + \dots + w_nx_n + w_0$$

Note que $h$ puede representarse como una multiplicación vectorial:
$$ h = w^Tx$$
donde el vector $w = [w_0,w_1,\dots,w_n]$ y el vector $x = [1,x_1,\dots,x_n]$


Finalmente, la función de activación normalmente empleada para problemas de clasificación binária (perceptron exitado o no), es la función sigmoide:
$$ \sigma(u) = \frac{1}{1 + e^{-u}}$$

```python
u = np.linspace(-10, 10, 1000)
sigmoide = 1/(1+np.exp(-u))
plt.grid()
plt.title("función sigmoide")
plt.xlabel("u")
plt.ylabel("sigmoide(u)")
plt.plot(u, sigmoide);
```

### Función de costo y entrenamiendo del Perceptron

Normalmente, en un dataset de entrenamiento se cuentan con pares de ejemplos de entrenamiento $(x, y)$, donde $x$ representan el ejemplo de entrada y $y$ representa el objetivo (target), siendo que para problemas de clasificación binaria, $y=1$ o $y=0$.

Suponga que estamos trabajando con un problema de clasificación binaria. Para poder penalizar nuestro perceptron (si clasifica bien o mal), tendremos que definir una función de costo, la cual es denotada por:
$$J({\bf{w}})=-\frac{1}{m}\sum_{j=1}^{m} \hat{y}\log(\sigma({\bf{w}}^T{\bf{x}}_j)) + (1-\hat{y})\log(1-\sigma({\bf{w}}^T{\bf{x}}_j))$$
$$J({\bf{w}})=-\frac{1}{m}\sum_{j=1}^{m} \hat{y}\log(\sigma(h)) + (1-\hat{y})\log(1-\sigma(h))$$

Donde $m$ representa el número de ejemplos de entrenamiento y $\hat{y}$ representa la clase predicha por el Perceptron.

Para entrenar el Perceptron se recurre al algoritmo del descenso del gradiente (<b>Gradient Descent</b>), el cual se basa en una regla iterativa para actualizar los pesos del Perceptron:
$${\bf{w}}(\tau) = {\bf{w}}(\tau-1) - \alpha\nabla J({\bf{w}})$$

$$w_i(\tau) = w_i(\tau-1) - \frac{\alpha}{m}\sum_{j=1}^{m} (\sigma({\bf{w}}^T{\bf{x}}_j) - \hat{y})x_{ji}$$

donde $\alpha$ representa la taza de aprendizaje, $x_{ji}$ es la $i$-th variable de entrada de el $j$-th ejemplo de entrenamiento.

```python
class perceptron:
    
    def __init__(self):
        """
        inicializar variables
        """
        self.error = []
        self.accuracy = []
    
    def sigmoide(self, u):
        """
        Computa la fución sigmoide
        """
        return 1/(1+np.exp(-u))

    def entrenar(self, X, y, lr=.001, iterations=10000):
        """
        Entrena el perceptron
        """
        m = X.shape[0]
        #se concatena el bias para cada ejemplo de entrenamiento
        X_train = np.concatenate((X, np.ones((X.shape[0],1))), axis=1)
        #se inicializan los pesos
        #w = np.array([30,-40,-120])
        self.w = np.random.normal(size=X_train.shape[1])
        #entrenamiento
        for i in range(iterations):
            #forward
            h = np.dot(X_train, self.w)
            y_hat = self.sigmoide(h.T)
            output_error = y_hat - y
            self.error.append(np.sum(abs(output_error))/m)
            #accuracy
            y_predict = (y_hat>.5)*1
            self.accuracy.append(np.mean(y_predict==y))
            #backward
            temp = np.dot(X_train.T, output_error.T)
            w_new = self.w - lr*temp/m
            self.w = w_new
        return self.w, self.error, self.accuracy
    
    def predecir(self, X):
        X_ = np.concatenate((X, np.ones((X.shape[0],1))), axis=1)
        h = np.dot(X_, self.w)
        return self.sigmoide(h)
```

### Probando nuestro Perceptron

- Para probar nuestro Perceptron utilizamos el conocido "iris dataset", solo que tomaremos unicamente dos de sus tres clases.
- Este dataset cuentra con tres clases, cada una con 50 ejemplos que representan metadatos de 3 flores diferentes.
- por cuestiones de aprendizaje únicamente tomamos 2 metadatos.

```python
from sklearn import datasets
#Cargamos el dataset
iris = datasets.load_iris()
X, y = iris.data, iris.target
#tomamos unicamente las dos primeras clases (100 ejemplos) y los dos primeros metadatos (columnas)
X_train = X[:100][:,:2]
y_train = y[:100]
fig, ax = plt.subplots(1,1)
ax.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap="Set2");
```

```python
#iniciamos la clase
p = perceptron()
w, error, accu = p.entrenar(X_train, y_train, lr=0.01, iterations=10000)
```

```python
plt.figure(figsize=(8,4))
plt.ylabel("eror")
plt.plot(error, label="error");
plt.plot(accu, label="Accuracy", color="orange");
plt.grid()
plt.xlabel("iteraciones");
plt.legend();
```

### Frontera de decisión construída por el Perceptron

```python
print("pesos aprendidos:", w)
x1 = np.linspace(4,7,100)
#ecuación de la línea (ver Anexos al final del notebook)
x2 = -(w[0]/w[1])*x1 - w[2]/w[1]
plt.grid()
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap="Set2")
plt.plot(x1, x2, label="frontera", color="black");
plt.legend();
```

```python
predict = (p.predecir(X_train) > 0.5)*1
print("precisión del perceptron:", sum(predict==y_train), "%")
```

### Conclusiones
- Note que un perceptron solo puede generar fronteras lineales.
- La inicialización de los pesos ayuda mucho para que el error pueda disminuir (perceptron converge)

### Perceptron con Tensorflow

```python
import tensorflow as tf

#creamos el perceptron
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=1,
                                input_shape=(X_train.shape[1],),
                                activation="sigmoid"))
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
              loss="binary_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
hist = model.fit(X_train, y_train, batch_size=32, epochs=120) #batch_size=32
```

```python
plt.figure(figsize=(10,3))
#Error
plt.subplot(1,2,1)
plt.ylabel("eror")
plt.plot(hist.epoch, hist.history["loss"], label="error");
plt.grid()
plt.legend()
plt.xlabel("epochs");
#Accuracy
plt.subplot(1,2,2)
plt.ylabel("Accuracy")
plt.grid()
plt.plot(hist.epoch, hist.history["accuracy"], label="Accuracy", color="orange");
plt.legend()
plt.xlabel("epochs");
```

#### Frontera de decisión

```python
w_tf = model.get_weights()
w1 = w_tf[0][0]
w2 = w_tf[0][1]
w0 = w_tf[1][0]

print("pesos aprendidos:", w1, w2, w0)
x1 = np.linspace(4,7,100)
#ecuación de la línea (ver notas al final)
x2 = -(w1/w2)*x1 - w0/w2   #x2 = -(w[0]/w[1])*x1 - w[2]/w[1]
plt.grid()
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap="Set2")
plt.plot(x1, x2, label="frontera", color="black");
plt.legend();
```

## Anexos

- Explicación de como pintar la frontera de decisión para un perceptron con 2 entradas ($x_1, x_2$)

$$ h = w_1x_1 + w_2x_2 + w_0 $$

hacemos cambio de variables ($x = x1, y = x2, b = w_0$)
$$ h = w_1x + w_2y + b $$

Note que si igualamos a cero la función $h$ obtenemos la ecuación general de la recta recta ($Ax + By + C = 0$).

Buscamos los interceptos:

- Intercepto en $y~(x=0)$:
$$ w_2y + b = 0$$
$$ y = \frac{-b}{w_2} $$

- Intercepto en $x~(y=0)$:
$$ w_1x + b = 0 $$
$$ x = \frac{-b}{w_1} $$

- Note que tenemos dos puntos en el plano 2D: $(0, -b/w_2)$ y $(-b/w_1, 0)$, con lo cual podemos calcular la pendiente de la recta:

$$ m = \frac{y_2-y_1}{x_2-x_1} $$

$$ m = \frac{ 0 - (-b/w_2)}{-b/w_1 - 0} $$


$$ m = -\frac{w_1}{w_2} $$

- con la pendiente calculada ya podemos calcular la función de la recta empleando la eciación general de la recta $y = mx + c$, donde $c=-b/w_2$

- Por lo tanto la función que representa la frontera de decisión es igual a:
$$ y = -\frac{w_1}{w_2}x -\frac{b}{w_2} $$
